```python
def fetch_playlist_by_id(sp, playlist_id):
    try:
        # Fetch the playlist details
        playlist = sp.playlist(playlist_id)
        
        # Print the playlist details
        print(f"Playlist Name: {playlist['name']}")
        print(f"Description: {playlist['description']}")
        print(f"Total Tracks: {playlist['tracks']['total']}")
        print(f"Owner: {playlist['owner']['display_name']}")
        
        # Print track details
        for track in playlist['tracks']['items']:
            track_info = track['track']
            print(f"Track Name: {track_info['name']}")
            print(f"Artist: {track_info['artists'][0]['name']}")
            print(f"Album: {track_info['album']['name']}")
            print('---')

    except spotipy.exceptions.SpotifyException as e:
        print(f"Spotify API error: {e}")
    except Exception as e:
        print(f"An error occurred: {e}")

# The Playlist ID you provided
playlist_id = '3Zyot4PjNacnll9W0f3wGl'

# Fetch and print the playlist details by ID
fetch_playlist_by_id(sp, playlist_id)
```

```python
tracks_info = []

for item in playlist['tracks']['items']:
    track_name = item['track']['name']
    artist_name = item['track']['artists'][0]['name']
    album_name = item['track']['album']['name']
    total_tracks = item['track']['album']['total_tracks']
    popularity = item['track']['popularity']
    duration_milliseconds = item['track']['duration_ms']
    release_date = item['track']['album']['release_date']
    
    tracks_info.append((track_name, artist_name, album_name, total_tracks, popularity, duration_milliseconds, release_date))

tracks_info
len(tracks_info)
```
